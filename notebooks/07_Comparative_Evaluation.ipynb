```python
import pandas as pd

# Load the results from the CSV files
baseline_results_df = pd.read_csv('C:/HGC/data/baseline_results.csv')
hgc_results_df = pd.read_csv('C:/HGC/data/hgc_results.csv')

# Concatenate the results into a single DataFrame for comparison
comparison_df = pd.concat([baseline_results_df, hgc_results_df], ignore_index=True)

print('--- Comparative Performance ---\n')
print('A lower perplexity indicates the model is more confident and accurate in its predictions.')
print('Our goal is for the HGC-Augmented model to have a lower perplexity than the baseline.')
display(comparison_df)
```

```python
import matplotlib.pyplot as plt

# Create a bar chart to visualize the comparison
plt.figure(figsize=(8, 6))
bars = plt.bar(comparison_df['model'], comparison_df['perplexity'], color=['skyblue', 'salmon'])
plt.ylabel('Perplexity (Lower is Better)')
plt.title('Performance Comparison: Baseline vs. HGC-Augmented Model')
plt.xticks(rotation=0)

# Add perplexity values on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, f'{yval:.2f}', va='bottom', ha='center')

plt.show()
```
